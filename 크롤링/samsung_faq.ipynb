```python
# pip install selenium webdriver-manager

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager

URL = "https://www.samsungfire.com/vh/page/VH.HPCS0611.do"

def check_samsung_structure():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--blink-settings=imagesEnabled=false")
    options.page_load_strategy = "eager"

    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )

    wait = WebDriverWait(driver, 10)
    driver.get(URL)
    wait.until(lambda d: len(d.find_elements(By.CSS_SELECTOR, ".accordion-wrap")) > 0)

    result = driver.execute_script("""
        function clean(t){
            return (t || '').replace(/\\s+/g, ' ').trim();
        }

        const root = document.querySelector('.accordion-wrap');
        if (!root) return null;

        const info = {
            tag: root.tagName,
            className: root.className,
            textPreview: clean(root.innerText).slice(0, 2000),
            childCount: root.children.length,
            children: [],
            buttons: [],
            headings: []
        };

        [...root.children].slice(0, 20).forEach((el, idx) => {
            info.children.push({
                index: idx,
                tag: el.tagName,
                className: el.className || '',
                text: clean(el.innerText).slice(0, 500),
                html: el.outerHTML.slice(0, 1000)
            });
        });

        [...root.querySelectorAll('button, a')].slice(0, 30).forEach((el, idx) => {
            info.buttons.push({
                index: idx,
                tag: el.tagName,
                className: el.className || '',
                text: clean(el.innerText),
                html: el.outerHTML.slice(0, 500)
            });
        });

        [...root.querySelectorAll('h1,h2,h3,h4,h5,strong,dt,dd,p,li,div')].slice(0, 40).forEach((el, idx) => {
            const txt = clean(el.innerText);
            if (txt) {
                info.headings.push({
                    index: idx,
                    tag: el.tagName,
                    className: el.className || '',
                    text: txt.slice(0, 300)
                });
            }
        });

        return info;
    """)

    driver.quit()

    if not result:
        print("accordion-wrap를 찾지 못함")
        return

    print("=" * 100)
    print("ROOT")
    print("tag:", result["tag"])
    print("class:", result["className"])
    print("childCount:", result["childCount"])

    print("\n" + "=" * 100)
    print("TEXT PREVIEW")
    print(result["textPreview"])

    print("\n" + "=" * 100)
    print("DIRECT CHILDREN")
    for c in result["children"]:
        print(f"\n[child {c['index']}] {c['tag']} class={c['className']}")
        print("TEXT :", c["text"])
        print("HTML :", c["html"])
        print("-" * 80)

    print("\n" + "=" * 100)
    print("BUTTON / LINK CANDIDATES")
    for b in result["buttons"]:
        print(f"\n[btn {b['index']}] {b['tag']} class={b['className']}")
        print("TEXT :", b["text"])
        print("HTML :", b["html"])
        print("-" * 80)

    print("\n" + "=" * 100)
    print("TEXT NODES")
    for h in result["headings"]:
        print(f"[{h['index']}] {h['tag']} class={h['className']} -> {h['text']}")

if __name__ == "__main__":
    check_samsung_structure()
```

```python
# pip install selenium webdriver-manager

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.samsungfire.com/vh/page/VH.HPCS0611.do"

def clean(text):
    return " ".join((text or "").split())

def get_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--blink-settings=imagesEnabled=false")
    options.page_load_strategy = "eager"

    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )

def extract_current_faqs(driver, category_name):
    rows = driver.execute_script("""
        function clean(t){
            return (t || '').replace(/\\s+/g, ' ').trim();
        }

        const out = [];
        document.querySelectorAll('dl.accordion-box').forEach(dl => {
            const q = clean(dl.querySelector('dt button')?.innerText);
            const a = clean(dl.querySelector('dd')?.innerText);
            if (q && a && !q.includes('{{')) {
                out.push({q, a});
            }
        });
        return out;
    """)

    result = []
    for row in rows:
        result.append({
            "category": category_name,
            "question": clean(row["q"]),
            "answer": clean(row["a"])
        })
    return result

def get_tab_texts(driver):
    return driver.execute_script("""
        function clean(t){
            return (t || '').replace(/\\s+/g, ' ').trim();
        }

        const els = [...document.querySelectorAll('button, a')];
        const tabs = els
            .map(el => clean(el.innerText))
            .filter(t => t && t.length < 20);

        return [...new Set(tabs)];
    """)

def click_tab_by_text(driver, tab_text):
    return driver.execute_script("""
        const targetText = arguments[0];

        function clean(t){
            return (t || '').replace(/\\s+/g, ' ').trim();
        }

        const els = [...document.querySelectorAll('button, a')];
        const target = els.find(el => clean(el.innerText) === targetText);

        if (target) {
            target.click();
            return true;
        }
        return false;
    """, tab_text)

def page_signature(driver):
    return tuple(driver.execute_script("""
        function clean(t){
            return (t || '').replace(/\\s+/g, ' ').trim();
        }
        return [...document.querySelectorAll('dl.accordion-box dt button')]
            .slice(0, 3)
            .map(el => clean(el.innerText));
    """))

def wait_until_tab_changed(driver, old_sig, timeout=6):
    end = time.time() + timeout
    while time.time() < end:
        new_sig = page_signature(driver)
        if new_sig and new_sig != old_sig:
            return True
        time.sleep(0.3)
    return False

def crawl_samsung_with_category():
    driver = get_driver()
    wait = WebDriverWait(driver, 10)

    driver.get(URL)
    wait.until(lambda d: len(d.find_elements(By.CSS_SELECTOR, "dl.accordion-box")) > 0)

    # 탭 이름 후보 확인
    raw_tabs = get_tab_texts(driver)

    # 화면에 보이는 실제 카테고리만 수동 필터
    categories = [
        t for t in raw_tabs
        if t in [
            "전체", "보험상품", "계약관리", "납입", "환급", "퇴직연금",
            "질병·상해 보상", "자동차보상", "재물·배상책임 보상",
            "대출", "혜택·서비스", "기업", "기타"
        ]
    ]

    print("카테고리 목록:", categories)

    all_rows = []
    seen = set()

    for category in categories:
        # 전체는 보통 중복이라 제외하고 싶으면 여기서 continue
        if category == "전체":
            continue

        old_sig = page_signature(driver)
        clicked = click_tab_by_text(driver, category)

        if not clicked:
            print(f"[{category}] 탭 클릭 실패")
            continue

        changed = wait_until_tab_changed(driver, old_sig, timeout=6)
        if not changed:
            # 같은 구조일 수도 있으니 그냥 추출 시도
            pass

        rows = extract_current_faqs(driver, category)

        added = 0
        for row in rows:
            key = (row["category"], row["question"], row["answer"])
            if key not in seen:
                seen.add(key)
                all_rows.append(row)
                added += 1

        print(f"[{category}] {added}건 추가 / 누적 {len(all_rows)}건")

    driver.quit()

    if not all_rows:
        print("추출 실패")
        return

    for i, row in enumerate(all_rows, 1):
        print(f"\\n[{i}]")
        print("카테고리:", row["category"])
        print("질문:", row["question"])
        print("답변:", row["answer"])
        print("-" * 100)

    print(f"\\n총 {len(all_rows)}건 추출 완료")

if __name__ == "__main__":
    crawl_samsung_with_category()
```
